```python
flat_data_names = [i for i in df.keys()]
flat_data = [df[i] for i in flat_data_names]
full = flat_data[0]
for i, j in enumerate(flat_data[1:]):
    name = flat_data_names[i+1]
    print(name)
    print(len(j['DBN']) - len(j['DBN'].unique()))
    join_type = 'inner'
    if name in ['sat_results', 'ap_2010', 'graduation']:
        join_type = 'outer'
    if name not in ['math_results']:
        full = full.merge(j, on='DBN', how=join_type)
full.shape
```

```python
cols = ['AP Test Takers ', 'Total Exams Taken',
        'Number of Exams with scores 3 4 or 5']
for i in cols:
    full[i] = full[i].convert_objects(convert_numeric=True)
full[i] = full[i].fillna(value=0)
```

```python
l =[]
for i in locat.values.tolist():
    start = i.find('(')
    end = i.find(')')
    l.append(i[start+1:end])
```

```python
lat = []
lon = []
for i in l[:]:
    start = i.find(',')
    lat.append(float(i[:start]))
    lon.append(float(i[start+1:]))
```

```python
import folium
from folium import plugins
```

```python
schools_map = folium.Map(location=[full['lat'].mean(),
                                 full['lon'].mean()],
                       zoom_start=10)
marker_cluster = plugins.MarkerCluster().add_to(schools_map)
for i, row in full.iterrows():
    folium.Marker([row['lat'], row['lon']], popup='{0}: {1}'.format(row['DBN'],
                                                                        row['school_name'])).add_to(marker_cluster)
schools_map.save('schools_map.html')
schools_map
```

```python
schools_heatmap = folium.Map(location=[full['lat'].mean(),
                            full['lon'].mean()],
                            zoom_start=10)
schools_heatmap.add_child(plugins.HeatMap([row['lat'],
                                             row['lon']] for i, row in full.iterrows()))
schools_heatmap.save('schools_heatmap.html')
schools_heatmap
```

```python
district_data = full.groupby('school_dist').mean()
district_data.reset_index(inplace=True)
district_data['school_dist'] = district_data['school_dist'].apply(lambda x: str(int(x)))
```

```python
def show_district_map(col):
    data1 = 'data/districts.geojson'
    districts_map = folium.Map(location=[full['lat'].mean(),
                                     full['lon'].mean()],
                          zoom_start=10)
    districts_map.add_child(folium.Choropleth(geo_data=data1,
                      data=district_data,
                      columns=['school_dist', col],
                      key_on='feature.properties.school_dist',
                      fill_color='YlGn',fill_opacity=0.8,
                      line_opacity=0.2))
    districts_map.save('schools_districts_{}.html'.format(col))
    return 'Ok'
```

```python
show_district_map('sat_score')
```
